# Project Title: Change in Mean Sea Level

#  a) Introduction and Dataset Description:

The Earth's oceans level is in a constant state of rising or falling, hence an observation must be done to prevent them from rising too high. Mean sea level is a crucial metric in oceanography, climate science, and environmental studies, as it defines midpoint level in a specific region; the half-point between the average high and low tides. This measure is vital for understanding various environmental and socio-economic impacts. Rising mean sea levels are a clear indicator of global warming, primarily caused by the melting of glaciers and ice sheets and the thermal expansion of seawater as it warms. Other than that, increased concentrations of greenhouse gases in the atmosphere trap more heat, leading to higher ocean temperatures and consequent sea level rise. Higher sea levels means there is a tendency to erode the coastal areas, thus affecting infrastructures, habitats, and coastal communities. Rising sea level might results to higher frequency and severity of storms and high tides which make the low-lying areas are vulnerable and exposed. Furthermore, increased levels of sea level leads to the intrusion of salt water in to the fresh water bodies that are used for agriculture and human consumption.


Rising sea level can submerge coastal habitats including mangroves and salt marshes that act as vital nurseries for several marine organisms. Floods accompanied by sea level changes may affect the dispersion of marine and coastal species thus posing a threat to marine bio-diversity and health. Economically, socio-economically related flooding and erosion can affect homes, businesses, and transportation infrastructure resulting in huge losses. Some of the possible impacts that can be observed in communities include displacement in small island countries and coastal cities due to sea level rise and more frequent flooding. Some of the prevention measures must be taken by the governments are building sea walls, flood defense and managed retreat to cope with this rising sea level matter.

The trend changes in mean sea level are measured using tide gauges which provide the long term records of mean sea level at particular coast stations, satellite altimetry that gives accurate measurements of changes in sea level from space and climate models which are used to predict future changes in sea level by using various emission scenarios of greenhouse gasses. Understanding the changes in mean sea level is crucial for developing effective climate policies and adaptation strategies. This research can help the policymakers in guiding actions to mitigate and adapt to the impacts of rising sea levels, aiming to protect vulnerable communities and ecosystems. In summary, "Change in Mean Sea Level" implies the assessment of the shifts in the mean sea-level according to regions, which illustrates the ways the overall climate system changes on the Earth.

#  f) Machine Learning Method based on Research Question in (b)

The objective of this analysis was to forecast sea level changes in the Indian Ocean and globally using historical data. We used two time series forecasting models, ***ARIMA and Prophet***, to predict future sea level changes.

Data Preprocessing:
The dataset was first filtered to include only measurements for the Indian Ocean and global data. Due to inconsistent time intervals in the raw data, we standardized these intervals by aggregating the data on a monthly basis, calculating the mean values for each month. A 'Date' column was created to facilitate time series analysis. The data was then split into training (95%) and testing (5%) sets to evaluate model performance.

Exploratory Data Analysis (EDA):
We visualized the training and testing data to observe trends and seasonal patterns in sea level changes over time. This helped us understand the underlying patterns in the data.

<br>

**Modeling and Forecasting:**

<br>

***ARIMA Model:***

**Hyperparameter Tuning:** We conducted a grid search over possible ARIMA parameters (p, d, q) to find the best combination that minimized the Akaike Information Criterion (AIC). The optimal parameters found were (2, 1, 2) for the Indian Ocean data and (2, 1, 2) for the global data, which corresponded to the lowest AIC values.

**Model Evaluation:** The performance of the ARIMA model was assessed using the test data. For the Indian Ocean data, the model provided accurate forecasts but was not explicitly evaluated using metrics in this summary. For the global data, the ARIMA model similarly showed good performance in capturing the overall trend.

<br>

**Prophet Model:**

**Data Preparation:** The data was reformatted to fit Prophet's requirements, with columns renamed appropriately.
Model Training: The Prophet model was trained on the training data.
Future Predictions: The model was used to make future predictions for the testing period. The forecast was plotted alongside the actual test data for comparison.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
import itertools
from prophet import Prophet
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df_model = df[df['Measure'] == 'Indian Ocean']
df_model.head(20)
```

#### The date is not consistent, sometimes the interval is in 10 days, sometimes 11. Thus, it will be aggregated on the mounthly bases to standardize the interval

```python
# group by year and month so that the time interval is standardized
df_model = df_model.groupby(['Year','Month']).agg({'Value':'mean'}).reset_index()
```

```python
df_model['Date'] = pd.to_datetime(df_model[['Year', 'Month']].assign(Day=1))
df_model
```

### Spliting Data into Train and Test

```python
train_idx = int(df_model.shape[0]*.95)

train = df_model.iloc[:train_idx]
test = df_model.iloc[(train_idx-1):]
```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(12, 6))
plt.plot(train['Date'], train['Value'], label='Train data')
plt.plot(test['Date'], test['Value'], label='Test data')
plt.xlabel('Date')
plt.ylabel('Sea Level Change (Millimeters)')
plt.title('Global Sea Level Changes Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

### Hyper parameter tuning

```python
# ARIMA Model
# Define the p, d, q parameters to take any value between 0 and 2
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

# Initialize variables to store the best parameters and the lowest AIC value
best_aic = float('inf')
best_pdq = None

# Grid search for the best parameters
for param in pdq:
    try:
        model = ARIMA(train['Value'], order=param)
        results = model.fit()
        if results.aic < best_aic:
            best_aic = results.aic
            best_pdq = param
    except:
        continue
```

```python
print("\nBest ARIMA Parameters:", best_pdq)
print("Lowest AIC:", best_aic)

# Fit the ARIMA model with the best parameters
model = ARIMA(train['Value'], order=best_pdq)
results = model.fit()
```

### Model training and forecast

```python
# Prophet Model
# Prepare the data for Prophet
prophet_df = train[['Date', 'Value']].rename(columns={'Date': 'ds', 'Value': 'y'})

# Initialize the Prophet model
prophet_model = Prophet()

# Fit the Prophet model
prophet_model.fit(prophet_df)

# Make future predictions with Prophet
future = prophet_model.make_future_dataframe(periods=test.shape[0], freq='M')  #
forecast = prophet_model.predict(future)
```

```python
# Plot the forecast
fig = prophet_model.plot(forecast)
plt.title('Prophet Model Forecast')
plt.plot(test['Date'], test['Value'], label='Test data',color='orange',alpha=.8)
plt.xlabel('Date')
plt.ylabel('Sea Level Change (Millimeters)')
plt.legend()
plt.show()
```

```python
#Model Evaluation
mean_squared_error(test['Value'],forecast['yhat'][(-test.shape[0]):])
```
