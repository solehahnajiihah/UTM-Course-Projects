Given a raw data of the record of the attendance for 15 employees in a company for a 14 days event. The exported raw data is recorded in following way:

(i) Read the “Attendance_RAW.txt” and save the data into a NumPy array.

```python
# WRITE YOUR CODE HERE

import numpy as np

path = 'Attendance_RAW.txt'

with open(path) as file:
    #data = file.readlines()
    data = [line.rstrip() for line in file]

arr1 = np.array(data, dtype = 'int32')
arr1
```

(ii) Shape the NumPy array to a structure where row represents the employees, column represents the working days.

```python
# WRITE YOUR CODE HERE

arr1 = arr1.reshape((15,14))
arr1
```

```python
# checking the shape, dimension and data type of arr1

print(arr1.shape)   # shape
print(arr1.ndim)    # dimension
print(arr1.dtype)   # data type
```

(iii) Construct a function that can calculate and return the list of average of attendance for each day.

```python

def avrg_attendance(workers):
    day_avg = workers.mean(axis=1)
    for i in range(15):
        #print("The average attendance for Day " + str(i) + " is " + str(day_avg[i])) # without rounding up 2 decimal places
        print(f"The average attendance for Day {i+1} is {day_avg[i]:.4f} ({day_avg[i]*100:.2f}%)")            # rounding up 2 decimal places
```

```python
# Testing the function

avrg_attendance(arr1)
```

(iv) Add a column with name ``total`` into the ``emp_attendance`` Data Frame that consists of the total no. of days the employee attended the event.

```python
# WRITE YOUR CODE HERE

emp_attendance['total'] = attendance.sum(axis=1)
emp_attendance
```

(v) Add a column with name ``rate`` into the ``emp_attendance`` Data Frame that consists of the attendance rate / percentage the employee throughout the 14 days.

```python
# WRITE YOUR CODE HERE

emp_attendance['rate'] = np.around(((attendance.sum(axis=1))/14)*100,2)
emp_attendance
```

(vi) Consider following rules in classify the attendance:<br>
        a.	Rate > 90% : Good <br>
        b.	60% < Rate < 90% : Moderate <br>
        c.	Rate < 60% : Bad <br>
    Based on the grouping rule above, get the status for each of the staff and save the status in a new column ``status`` in the Data Frame.

```python
# WRITE YOUR CODE HERE

def emp_status(row):
    if row['rate'] > 90:
        stat = 'Good'
    elif (row['rate'] > 60) & (row['rate'] < 90):
        stat = 'Moderate'
    else:
        stat = 'Bad'
        
    return stat
```

```python
emp_attendance['status'] = emp_attendance.apply(emp_status, axis=1)
emp_attendance
```

```python
# Another way: using list comprehension

emp_attendance["Status"] = ["Bad" if s < 60 else "Moderate" if ((s>60) & (s<90)) else "Good" for s in emp_attendance["rate"]]
emp_attendance
```
