## Question 3

Given two dataset (`Death Male.xlsx` and `Death Female.xlsx`) that contains the data of recorded death in five states in Malaysia from 2011 to 2018.

**Task A**

(i)	Read the `Death Male.xlsx` and store it in a DataFrame named `death_male`. Group the data by `Year` and name the result as `dm_by_year` and plot a pie chart.


```python
death_male = pd.read_excel('Death Male.xlsx')
death_male
```

```python
dm_by_year = death_male.groupby(death_male['Year'])['Number of death'].sum()
dm_by_year
```

```python
figQ3_1 = plt.figure(figsize=(10, 5))
figQ3_1 = dm_by_year.plot(kind='pie', title= 'Number of Death by Year', autopct= '%1.2f%%', textprops={'fontsize':9})
figQ3_1.set_ylabel('')
```

(ii) Read the `Death Female.xlsx` and store it in a DataFrame named `death_female`. Group the data by `State` and name the result as `df_by_state` and plot an appropriate chart.

```python
death_female = pd.read_excel('Death Female.xlsx')
death_female
```

```python
df_by_state = death_female[['Number of death']].groupby(death_male['State']).sum().sort_values(by=['Number of death'], ascending=False)
df_by_state
```

```python
figQ3_2 = plt.figure(figsize=(10, 5))
figQ3_2 = df_by_state.plot(kind='barh', title= 'Number of Death by State')
figQ3_2.set_xlabel('Number of Death')
```

**Task 2**

(i)	  Concatenate both `death_male` and `death_female` DataFrame and name the new DataFrame as `death`.	<br>
(ii)  Group the `death` by `State` and name the result as `by_state`.	<br>
(iii) Group the `death` by `Year` name the result as `by_year`.	<br>
(iv)  Create a chart as shown below with figure size of 12 inches × 6 inches and save it as Death in 2011-2018.png with dpi value 200.


<br>
<img src="Fig1.png", style="width: 750px";>

```python
# Concatenate both death_male and death_female DataFrame and name the new DataFrame as death.

death = pd.concat([death_male, death_female])
death
```

```python
# Group the death by State and name the result as by_state

by_state = death.groupby('State')[['Number of death']].sum()
by_state
```

```python
# Group the death by Year name the result as by_year

by_year = death.groupby('Year')[['Number of death']].sum()
by_year
```

```python
#  Create a chart as shown below with figure size of 12 inches × 6 inches and save it as Death in 2011-2018.png with dpi value 200

figQ3_3, axes = plt.subplots(1,2,figsize=(12,6))

by_state.plot(ax=axes[0], kind='bar', color='red', legend=False, title= 'Death from 2011 to 2018 by State') 
by_year.plot(ax=axes[1], kind='barh', color='blue', legend=False, title= 'Death from 2011 to 2018 by Year') 

axes[0].set_title('Death from 2011 to 2018 by State', fontsize=10)
axes[1].set_title('Death from 2011 to 2018 by Year', fontsize=10)

axes[0].set_ylabel('Number of Death', fontsize=10)
axes[1].set_xlabel('Number of Death', fontsize=10)

figQ3_3.tight_layout()
```

```python
# saving graph

figQ3_3.savefig("Death in 2011-2018.png", dpi=200)
```
